```python
from langchain.llms import OpenAI
```

```python
import os
```

```python
openai_api=os.getenv('OPENAI_API_KEY')
```

```python
llm=OpenAI(temperature=0.6,openai_api_key=openai_api)
```

```python
text = "what is the capitol of India"
print(llm.predict(text))
```
